## Introduction

Suicide is a complex and sensitive issue that affects individuals and communities worldwide. Understanding the factors influencing suicide rates is crucial for developing effective prevention strategies. This Python project delves into the analysis of suicide rates data, aiming to provide valuable insights into patterns, trends, and potential contributing factors.

## Project Objective

The objective of the project  is to analyze and understand the factors influencing suicide rates across different countries. This includes exploring the relationship between suicide rates and various socio-economic indicators such as Human Development Index (HDI), GDP, and other demographic factors. The goal is to identify patterns, trends, and potential correlations that can provide insights into the underlying causes of suicide and inform public health and policy interventions.

## Hypothesis

*1.* The global suicide rate has changed significantly from 1985 to 2021.

*2.* Certain age groups may have higher suicide rates, which could be influenced by factors such as social pressures, mental health issues, and economic conditions. For example, middle-aged individuals might have higher rates due to potential mid-life crises, financial burdens, or family responsibilities.

*3.* Different generations may experience varying suicide rates, potentially due to the unique challenges and societal changes they have faced during their lifetimes. For instance, older generations might have higher rates due to isolation, health issues, or lack of mental health awareness in their youth.

*4.* Economic factors such as GDP per capita could correlate with suicide rates, where countries with lower economic performance might have higher rates due to increased financial stress and lower access to mental health resources.

*5.* Gender differences in suicide rates may be significant, with males potentially having higher rates due to societal norms around masculinity and seeking help for mental health issues.

```python
# Graph of number of null values
def visualize_null(dataframe, percentage=True):
    # Calculates the number of null values per column and sorts them in descending order
    null_series = dataframe.isna().sum().sort_values(ascending=False)

    # If the percentage parameter is True, converts the number of nulls to a percentage
    if percentage:
        null_series = null_series / len(dataframe) * 100

    # Create a bar chart with the columns on the x-axis and the number or percentage of nulls on the y-axis
    plt.bar(x=null_series.index, height=null_series.values)

    # Set the y-axis label based on whether it is percentage or quantity
    if percentage:
        plt.ylabel('Percentage of missing values')
    else:
        plt.ylabel('Amount of missing values')

    # Set the title and labels of the axis
    plt.title('Missing values')
    plt.xlabel('Column variables')
    plt.xticks(rotation=90) 

    # Show the graph
    plt.show()

# Call the function with a specific DataFrame
visualize_null(suicide_df)
```

```python
# Function to find non-convertible integers in a given series
def find_non_convertible_integer(series):
    
    # Initialize an empty set to store non-convertible items
    non_convertibles = set()
    
    # Loop through each item in the series
    for item in series:
        try:
            int(item)  # Try converting to integer
        except ValueError:
            non_convertibles.add(item)  # Add to set if conversion fails
    # Return the set of non-convertible items
    return non_convertibles

find_non_convertible_integer(suicide_df['suicides_no'])
```

I used a **regression analysis** to further understand the relationship between generation and suicide rates, controlling for other factors

```python
import statsmodels.api as sm

# Encode the 'generation' column to prepare for regression analysis
label_encoder = LabelEncoder()
suicide_df['generation_encoded'] = label_encoder.fit_transform(suicide_df['generation'])

# Prepare the independent variables (X) and the dependent variable (y)
X = suicide_df[['generation_encoded', 'year', 'population', 'HDI for year', 'gdp_per_capita']]
X = sm.add_constant(X) # adding a constant for the intercept
y = suicide_df['suicides/100k pop']

# Handle missing values by imputing with the mean
X = X.fillna(X.mean())

# Perform the regression analysis
model = sm.OLS(y, X).fit()

# Get the summary of the regression
model_summary = model.summary()
model_summary
```

The findings of the regression analysis show that there is no statistically significant relationship between the encoded generation variable and the suicide rates per 100k people (p-value: 0.785). This shows that the generation may not be a reliable indicator of suicide rates on its own when year, population, HDI, and GDP per capita are taken into account.

The model's R-squared value is 0.015, indicating that it can account for just 1.5% of the variation in suicide rates. This indicates that there might be more variables that are not taken into account by the model but could have a big influence on suicide rates.
